## Section 1 - Obsidian Journal Frontmatter (Properties) Collection

```python
# Frontmatter Documentation https://pypi.org/project/python-frontmatter/
# Primary collection module is pyfront

vault_path = r"C:\Users\emerg\OneDrive\Documents\_Mimir Cyber\Coding\Personal Projects\obsidian-notes\Journal\Daily"

def create_nested_dictionary_from_vault(directory):
    daily_dic = {}

    
    for filename in os.listdir(directory):
        file_path = os.path.join(directory, filename)


        if os.path.isfile(file_path):
            # Remove the file extension to get the key
            date_key = os.path.splitext(filename)[0]
            

            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    post = pyfront.loads(file.read())
                    daily_dic[date_key] = post.metadata
            except (UnicodeDecodeError, ScannerError) as e:
                print(f"Error in file {filename}: {e}")
                continue  # Skips the file causing the error
                # For complete data, go to the skipped files and fix the errors


    return daily_dic

result_nested_dictionary = create_nested_dictionary_from_vault(vault_path)

pprint(result_nested_dictionary["2023-12-01"]) # Test print of the nested dictionary to make sure it 
```

## Section 2 - Timular Data Extraction via API

```python
#### Timeular API Documentation https://developers.timeular.com/#832677a3-73d5-4bde-85dc-40af433414c1 ####

# Secrets will load from ".env"
load_dotenv()
time_api = os.getenv("TIMEULAR_API_KEY")
time_secret = os.getenv("TIMEULAR_API_SECRET")
time_token = os.getenv("TIMEULAR_API_TOKEN")


# Timeular requires you to have an ApiKey and ApiSecret to get your ApiToken. This uses ApiKey and ApiSecret to make the required API Call.
token_payload = json.dumps({
    "apiKey": time_api,
    "apiSecret": time_secret,
})
token_headers = {
    'Content-Type': 'application/json'
}
token_url = "https://api.timeular.com/api/v3/developer/sign-in"
timeular_response_token = requests.request("POST", token_url, headers=token_headers, data=token_payload)
timeular_response_token.raise_for_status()
os.getenv('TIMEULAR_API_TOKEN', timeular_response_token.text)


# With your ApiToken, you can now export data for the dates in question. 
activities_headers = {
  'Authorization': f'Bearer {time_token}'
}
activities_payload = {}
activites_list_url = "https://api.timeular.com/api/v3/report/2023-01-01T00:00:00.000/2023-12-31T23:59:59.999?timezone=America/Chicago"
timeular_response_activites = requests.request("GET", activites_list_url, headers=activities_headers, data=activities_payload)
timeular_response_activites.raise_for_status()


acitivities_text = timeular_response_activites.text

# Setting output to be same directory as jupyter file
script_dir = os.getcwd()
output_file_path = os.path.join(script_dir, 'timeular_data.csv')

# Making the CSV readable for Jupyter manipulation
df = pd.read_csv(StringIO(acitivities_text))
df = df[['Space', 'StartDate', 'Activity', 'Duration', 'Tags']]
df.to_csv(output_file_path, index=False)
```

# Section 3b - Timular Visualization

```python
df = pd.read_csv('timeular_data.csv')

unique_labels_space = df['Space'].unique()
unique_labels_activity = df['Activity'].unique()
unique_labels_tags = df['Tags'].str.split().explode().unique()

# Concatenate all unique labels
all_unique_labels = list(unique_labels_space) + list(unique_labels_activity) + list(unique_labels_tags)

# Create a Sankey diagram
fig = pgo.Figure(data=[
    pgo.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color='black', width=0.5),
            label=all_unique_labels
        ),
        link=dict(
            source=df['Space'].map(lambda x: all_unique_labels.index(x)),
            target=df['Activity'].map(lambda x: all_unique_labels.index(x)),
            value=df['Duration']
        )
    ),
    # Add a link from 'Activity' to 'Tags'
    pgo.Sankey(
        link=dict(
            source=df['Activity'].map(lambda x: all_unique_labels.index(x)),
            target=df['Tags'].str.split().explode().map(lambda x: all_unique_labels.index(x)),
            value=df['Duration']
        )
    )
])

fig.show()
```

```python
import seaborn as sns
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Assuming you already have the aggregated_df and breakdown_df from previous code

# Convert 'Duration' to hours and minutes
aggregated_df['Duration_hours'] = aggregated_df['Duration_seconds'] // 3600
aggregated_df['Duration_minutes'] = (aggregated_df['Duration_seconds'] % 3600) // 60

# Create a Sankey diagram for "Off-work" space
off_work_data = aggregated_df[aggregated_df['Space'] == 'Off-Work Space']
fig_off_work = make_subplots(rows=1, cols=1, specs=[[{'type': 'sankey'}]])

# Flatten the 'Tags' column in off_work_data
off_work_data.loc[:, 'Tags'] = off_work_data['Tags'].apply(lambda x: x if isinstance(x, list) else [x])
tags_flat_off_work = off_work_data['Tags'].explode()

# Ensure tags are unique
tags_flat_off_work = tags_flat_off_work.unique()

# Combine 'Activity' and 'Tags' for node labels
node_labels_off_work = off_work_data['Activity'].unique().tolist() + tags_flat_off_work.tolist()

# Assign indices for nodes
node_indices_off_work = {label: index for index, label in enumerate(node_labels_off_work)}

# Assign indices for source nodes
link_source_indices_off_work = [node_indices_off_work[activity] for activity in off_work_data['Activity']]

# Assign indices for target nodes
link_target_indices_off_work = [len(node_indices_off_work) + index for index, tag in enumerate(tags_flat_off_work)]

link_values_off_work = off_work_data['Duration_seconds']

fig_off_work.add_trace(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color='black', width=0.5),
        label=node_labels_off_work,
    ),
    link=dict(
        source=link_source_indices_off_work,
        target=link_target_indices_off_work,
        value=link_values_off_work,
    )
))

fig_off_work.update_layout(
    title_text="Activity Duration Sankey Diagram (Off-work)",
    font_size=10,
    width=800,  # Adjust the width as needed
    height=600,  # Adjust the height as needed
)

# Create a Sankey diagram for "Work" space
work_data = aggregated_df[aggregated_df['Space'] == 'Work-Space']
fig_work = make_subplots(rows=1, cols=1, specs=[[{'type': 'sankey'}]])

# Flatten the 'Tags' column in work_data
work_data.loc[:, 'Tags'] = work_data['Tags'].apply(lambda x: x if isinstance(x, list) else [x])
tags_flat_work = work_data['Tags'].explode()

# Ensure tags are unique
tags_flat_work = tags_flat_work.unique()

# Combine 'Activity' and 'Tags' for node labels
node_labels_work = work_data['Activity'].unique().tolist() + tags_flat_work.tolist()

# Assign indices for nodes
node_indices_work = {label: index for index, label in enumerate(node_labels_work)}

# Assign indices for source nodes
link_source_indices_work = [node_indices_work[activity] for activity in work_data['Activity']]

# Assign indices for target nodes
link_target_indices_work = [len(node_indices_work) + index for index, tag in enumerate(tags_flat_work)]

# Convert duration to hours and minutes
work_data.loc[:, 'Duration_hours'] = work_data['Duration_seconds'] // 3600
work_data.loc[:, 'Duration_minutes'] = (work_data['Duration_seconds'] % 3600) // 60

link_values_work = work_data['Duration_seconds']

fig_work.add_trace(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color='black', width=0.5),
        label=node_labels_work,
    ),
    link=dict(
        source=link_source_indices_work,
        target=link_target_indices_work,
        value=link_values_work,
    )
))

fig_work.update_layout(
    title_text="Activity Duration Sankey Diagram (Work)",
    font_size=10,
    width=1280,  # Adjust the width as needed
    height=1024,  # Adjust the height as needed
)

# Show the Sankey diagrams
fig_off_work.show()
fig_work.show()
```
